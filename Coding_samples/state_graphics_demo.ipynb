```python
import pandas
from plotnine import *
```

```python
state_data = pandas.read_csv("https://raw.githubusercontent.com/Al-Arsenault201/Fall_2021_Project_3/main/11-11-2021.csv")
state_data.head(15)
```

```python
state_plot = ggplot(state_data, aes(x="Deaths", y="Total_Test_Results")) + geom_point()
state_plot
```

```python
state_plot = ggplot(state_data, aes(x="Deaths", y="Total_Test_Results", color="Province_State")) + geom_point()
state_plot
```

```python
state_plot = ggplot(state_data, aes(x="Deaths", y="Total_Test_Results")) + geom_col()
state_plot
```

```python
state_plot = ggplot(state_data, aes(x='Province_State', y="Deaths")) + geom_col()
state_plot
```

```python
states_I_want = ["Alabama","Alaska","Arizona","Arkansas"]
new_list = state_data[state_data["Province_State"].isin(states_I_want)]
new_list
```

```python
state_plot = ggplot(new_list, aes(x='Province_State', y="Deaths")) + geom_col()
state_plot
```
