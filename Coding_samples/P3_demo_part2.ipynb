```python
DATA_DIRECTORY = "/Users/alfredarsenault/Documents/Fall_21_p3_data_directory"
import os  #import a bunch of operating-system specific routines
```

```python
database = []
```

```python
os.chdir(DATA_DIRECTORY)
for file in os.listdir(DATA_DIRECTORY):
#    print(file)
    if ".csv" in file:
        with open (file, 'r') as infile:
            infile.readline()  #throws away the header line that I don't care about
            data = infile.read()  # rest of the file as a string
            data_list = data.split('\n') # split into a list of lines
            for i in range(len(data_list)):   # for every line in the file
                data_list[i] = data_list[i].split(',')
                data_list[i].append(int(file[3]+file[4]))
                database.append(data_list[i])
            
```

```python
print(database)
```

```python
better_db = []
for i in range(len(database)): # for every row
    new_row = []
    new_row.append(database[i][0])
    new_row.append(database[i][5])
    new_row.append(database[i][6])
    new_row.append(database[i][10])
    new_row.append(database[i][11])
    new_row.append(database[i][13])
    new_row.append(database[i][18])
    
    #I have now built a new row, and can append this row to the new database
    
    better_db.append(new_row)
    
```
