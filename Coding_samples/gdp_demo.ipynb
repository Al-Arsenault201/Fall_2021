```python
import pandas
from plotnine import *
```

```python
data = pandas.read_csv("https://raw.githubusercontent.com/swcarpentry/r-novice-gapminder/gh-pages/_episodes_rmd/data/gapminder-FiveYearData.csv")
data.head()
```

```python
myplot = ggplot(data,aes(x="lifeExp",y="gdpPercap",color='continent')) + geom_point()
myplot
```

```python
myplot2 = ggplot(data,aes(x="lifeExp",y="gdpPercap")) + geom_col()
myplot2
```

```python
myplot = ggplot(data,aes(x="lifeExp",y="gdpPercap")) + geom_point(size= 0.5)
myplot
```

```python
myplot = ggplot(data,aes(x="lifeExp",y="gdpPercap", size = 0.5)) + geom_point()
myplot
```

```python
multi_line_plot =(ggplot(data, aes(x='year', y='lifeExp', group='country', color='continent')) +
        geom_line(alpha = 0.5)
)
multi_line_plot
```
